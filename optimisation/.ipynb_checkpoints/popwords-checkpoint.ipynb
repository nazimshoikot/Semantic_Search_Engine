```python
import collections
import os
```

```python
from collections import Counter 
```

```python
directory = '/Users/adyasingh/Desktop/Semantic-Search-Engine/text_files/txt'
data = []
```

```python
for file in os.listdir(directory):
#     print(file)
    f = open(os.path.join(directory, file), "r", encoding="utf8")
    fileContent = f.read()
    data.append(fileContent.lower())
    f.close()
```

```python
data_str = ''.join(data)
data_str = data_str.lower()
split_it = data_str.split() 
```

```python
  
# Pass the split_it list to instance of Counter class. 
Counter = Counter(split_it) 
```

```python
  
# most_common() produces k frequently encountered 
# input values and their respective counts. 
most_occur = Counter.most_common(3000) 
  
print(most_occur) 
```

```python
popular_words = open("popular_words.txt","w+")
```

```python
for word in most_occur: 
#     popular(word[0])
    popular_words.write(word[0]+"\n")
popular_words.close()
```
